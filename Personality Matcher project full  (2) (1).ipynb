```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import KFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
import pandas as pd

# 1. Load dataset
df = pd.read_excel("mahabharata_200_warriors.csv.xlsx")

# 2. Combine relevant text columns
text_cols = ['Role/Title','Allegiance','Primary Weapon(s)',
             'Special Skills/Boons','Notable Contributions/Events']
df['combined'] = df[text_cols].astype(str).agg(' '.join, axis=1)

# 3. Target (Allegiance is easier than Name)
X_text = df['combined']
y = df['Allegiance']  

# 4. Pipeline: TF-IDF + Random Forest
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=1000)),
    ('rf', RandomForestClassifier(random_state=42))
])

# 5. KFold CV
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(pipeline, X_text, y, cv=kf)

# 6. Results
print("Individual fold scores:", scores)
print("Average CV Accuracy:", scores.mean())
```

#### TF-IDF tells the model which words matter most in describing a specific document and which words are just filler.
So after applying TfidfVectorizer(max_features=1000), each warrior‚Äôs text becomes a 1000-dimensional vector ‚Äî one number for each important word.

#### K-Fold is a model evaluation technique used to check how well our model generalizes to unseen data.
üîπ The Problem It Solves

If we train our model once on a single train/test split, our accuracy might depend on how the data was split ‚Äî which can be misleading.

K-Fold Cross Validation avoids this problem.

```python
def expand_text_min_words(text, min_words=100):
    
    words = text.split()
    if len(words) >= min_words:
        return text  # already long enough
    else:
        
        needed = min_words - len(words)
        # simple expansion: original text + motivational/descriptive filler
        filler = (" This character stands as a symbol of bravery, virtue, and moral duty "
                  "in the Mahabharata, reflecting courage, skill, compassion, loyalty and deep spiritual wisdom.")*5
        filler_words = filler.split()
        extra = filler_words[:needed]
        return text + " " + " ".join(extra)


df['expanded_text'] = df['combined_text'].apply(lambda x: expand_text_min_words(x, min_words=100))
```

The goal of this function is to make sure that each text entry in your dataset has at least 100 words ‚Äî which can be useful when training text-based models like TF-IDF or language models that perform better with longer, richer text.

This function ensures every warrior‚Äôs description has at least 100 words, by appending meaningful filler text if it‚Äôs too short ‚Äî helping text-based machine learning models train more effectively.

```python
import re

def clean_and_simple_grammar(text):
    # 1. Remove punctuations
    text = re.sub(r'[^\w\s]', '', text)
    
    # 2. Fix some basic grammar words (optional, simple examples)
    # Ye sirf readability ke liye, zyada structure change nahi karega
    text = re.sub(r'\bMale\b', 'Male', text)
    text = re.sub(r'\bFemale\b', 'Female', text)
    
    # 3. Lowercase + extra spaces remove
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# Apply on expanded_text
df['clean_text'] = df['expanded_text'].apply(clean_and_simple_grammar)

# Check result
print(df[['Name', 'clean_text']].head(5))
```

This function ‚Äî clean_and_simple_grammar(text) ‚Äî is used to clean the text data so that it‚Äôs simpler, more uniform, and machine-readable before using it for Natural Language Processing (NLP) or text classification.

Essentially, it‚Äôs preparing your ‚Äúexpanded‚Äù warrior descriptions for model input (like TF-IDF or other embeddings).

#### This removes all punctuation marks like:

Commas (,)

Periods (.)

Exclamation marks (!)

Quotes (" ")

\w means word characters (letters, digits, underscore)

\s means whitespace

So [^...] removes everything except words and spaces.

```python
import re

def fix_encoding_and_clean(text):
    # Common replacements
    replacements = {
        "√¢‚Ç¨‚Ñ¢": "'",   # apostrophe
        "√¢‚Ç¨‚Äú": "-",   # dash
        "√¢‚Ç¨≈ì": '"',   # opening quote
        "√¢‚Ç¨ÔøΩ": '"',   # closing quote
        "√¢‚Ç¨": '"',    # misc quotes
    }
    
    for k, v in replacements.items():
        text = text.replace(k, v)
    
    # Remove extra punctuations but keep apostrophes and dashes
    text = re.sub(r'[^\w\s\'-]', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# Apply to dataframe
df['clean_text'] = df['expanded_text'].apply(fix_encoding_and_clean)

# Check last row example
print(df.loc[130, 'clean_text'])
```

```python
def clean_text_func(text):
    if not isinstance(text, str):
        return ""
    text = re.sub(r'[^A-Za-z\s]', ' ', text)
    text = re.sub(r'\b[A-Za-z]\b', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['clean_text'] = df['clean_text'].apply(clean_text_func)

# Include 7 main personalities
main_chars = ['Yudhishthira', 'Bhima', 'Arjuna', 'Duryodhana', 'Karna', 'Shakuni', 'Krishna']
df = df[df['Name'].isin(main_chars)].reset_index(drop=True)
print(f"‚úÖ Loaded {len(df)} main characters.\n")
```

```python
questions = {
    "Q1": "When facing a major conflict, your first instinct is to...\n"
          "(A) Find a moral and just solution (Dharma).\n"
          "(B) Use my strength to solve it directly.\n"
          "(C) Analyze the situation and execute a precise, skilled plan.\n"
          "(D) Do whatever it takes to win and assert my claim.\n"
          "(E) Stay loyal to my friends who have supported me, no matter the cost.\n"
          "(F) Devise a clever or cunning strategy.\n"
          "(G) See the 'bigger picture' and guide events from a higher perspective.\n"
          "Your choice (A-G): ",

    "Q2": "What do you value most in yourself?\n"
          "(A) My integrity and truthfulness.\n"
          "(B) My physical power and passion.\n"
          "(C) My discipline and focused skill.\n"
          "(D) My ambition and determination.\n"
          "(E) My loyalty and generosity.\n"
          "(F) My intelligence and strategic mind.\n"
          "(G) My wisdom and ability to understand others.\n"
          "Your choice (A-G): ",

    "Q3": "A difficult personal dilemma arises. You are most likely to feel...\n"
          "(A) Burdened by the weight of the 'right' choice.\n"
          "(B) Impatient and ready for action.\n"
          "(C) Conflicted, but focused on my duty.\n"
          "(D) Deeply jealous or angry at the perceived injustice.\n"
          "(E) A tragic sense of loyalty to a difficult cause.\n"
          "(F) Amused, seeing it as a game to be won.\n"
          "(G) Calm and detached, understanding the role I must play.\n"
          "Your choice (A-G): ",

    "Q4": "Your greatest strength is...\n"
          "(A) My unwavering virtue.\n"
          "(B) My immense physical power.\n"
          "(C) My unmatched expertise in my field (e.g., archery, arts).\n"
          "(D) My powerful will and leadership.\n"
          "(E) My profound sense of gratitude and loyalty.\n"
          "(F) My ability to out-think my opponents.\n"
          "(G) My cosmic wisdom and charm.\n"
          "Your choice (A-G): ",

    "Q5": "People might criticize you for being...\n"
          "(A) Too rigid, passive, or naive.\n"
          "(B) Too rash or aggressive.\n"
          "(C) Prone to ego or moral doubt.\n"
          "(D) Envious and greedy.\n"
          "(E) Blinded by your loyalties.\n"
          "(F) Manipulative.\n"
          "(G) Detached, mysterious, or a rule-breaker.\n"
          "Your choice (A-G): "
}
```

```python
descriptions = {
    "Yudhishthira": "You are 'The Just.' You value dharma, truth, and morality above all. "
                    "You are righteous and calm, but this can make you seem passive or rigid at times.",

    "Bhima": "You are 'The Strong.' You are a person of action, passion, and immense power. "
             "You are fiercely protective, but your quick temper can sometimes be your downfall.",

    "Arjuna": "You are 'The Skilled.' You are focused, disciplined, and a master of your craft. "
              "You are driven by duty but can be prone to self-doubt and moral dilemmas.",

    "Duryodhana": "You are 'The Ambitious.' You are a powerful leader with an unbreakable will. "
                  "You are determined to get what you believe is yours, but this can stem from deep envy.",

    "Karna": "You are 'The Loyal.' You are generous, powerful, and defined by your loyalty to those who helped you. "
             "You have a tragic sense of honor, often sticking to your cause even when you know it's flawed.",

    "Shakuni": "You are 'The Cunning.' You are highly intelligent, strategic, and see life as a game to be won. "
               "You are a master manipulator, driven by a long-held grudge.",

    "Krishna": "You are 'The Guide.' You are wise, charming, and see the 'bigger picture' that others miss. "
               "You operate on a higher level, understanding that sometimes rules must be bent for a greater good."
}
```

```python
print("üîπ Loading sentence transformer model...")
embed_model = SentenceTransformer('all-MiniLM-L6-v2')

X = embed_model.encode(df["clean_text"], show_progress_bar=False)
y = df["Name"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LogisticRegression(max_iter=1000, class_weight='balanced')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
def get_user_answers():
    answers = []
    valid = set(list("ABCDEFG"))
    for qkey in questions:
        while True:
            ans = input(questions[qkey]).strip().upper()
            if ans in valid:
                answers.append(ans)
                break
            else:
                print("Invalid input. Please enter A‚ÄìG.")
    return answers
```

```python
option_to_character = {
    'A': 'Yudhishthira',
    'B': 'Bhima',
    'C': 'Arjuna',
    'D': 'Duryodhana',
    'E': 'Karna',
    'F': 'Shakuni',
    'G': 'Krishna'
}
```

```python
def predict_personality():
    print("\n‚ú® Answer the 5 questions to discover your Mahabharata personality ‚ú®\n")
    answers = get_user_answers()
    selected_characters = [option_to_character[a] for a in answers]

    # Combine chosen personalities' traits into one text
    combined_text = " ".join(df[df["Name"].isin(selected_characters)]["clean_text"].tolist())
    user_embed = embed_model.encode([combined_text])
    pred = model.predict(user_embed)[0]
    conf = max(model.predict_proba(user_embed)[0]) * 100

    print("\n========================================")
    print("           YOUR MAHABHARATA SELF         ")
    print("========================================\n")
    print(f"üß≠ You align most with: {pred.upper()}!")
    print(f"(AI Confidence: {conf:.2f}%)\n")
    print(textwrap.fill(descriptions[pred], width=80))
    print("\n========================================\n")
```

```python
if __name__ == "__main__":
    print("üåü MAHABHARATA PERSONALITY MATCHER üåü")
    print("Now using 7 characters and deep semantic AI!\n")
    predict_personality()
```
